```python
#Librerias
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Definición de variables
N= 10000   #cantidad de observaciones
Pi= 50000       #población inicial
dp= 1.0/800  #estimado de decaimiento

#Matriz Inicial
a= np.zeros((N, Pi))

#Observacion inicial
a[0, : ]=1
pobl=[Pi]
#iterador de numeros aleatorios para cada partícula
i=1
while sum(a[i-1, :])>0 and i<N:
  x=np.random.random(Pi)
  y=np.where(x<= dp)
  #print(y)
  a[i, : ]=a[i-1, : ]
  a[i,y] = 0              #
  po=sum(a[i,:])
  pobl.append(po)
  #print(po)
  i+=1
 
#print(pobl)
time=sum(a[: , :])    #Vector de tiempos de vida individual
#print(a)
#print(time)
```

```python
#graficas 
#muestra de decaimiento exponencial
plt.figure(figsize=(10, 10))
plt.subplot(2, 1, 1)
plt.plot(time)
plt.ylabel('Tiempo de Vida')
plt.xlabel('Isótopo')
plt.title('Tiempo de Vida Individual')

 
#muestra de decaimiento exponencial
plt.figure(figsize=(8, 8))
plt.subplot(2, 1, 1)
plt.plot(pobl)
plt.ylabel('Concentración de Isótopos')
plt.xlabel('Tiempo')
plt.title('Decaimiento del isótopo')
```

```python
#vida promedio
meanTime=10*sum(time)/len(time)
decayProbability=1.0/meanTime
halflifeTime= np.log(2)/decayProbability
print(f"El promedio de vida de los isótopos es: {meanTime} años ")
print(f"La probabilidad de decaimiento es: {decayProbability} 1/años ")
print(f"La vida media de los isótopos es: {halflifeTime} años ")
```

```python
def Edad(radio,decay):
  t=-np.log(radio)/decay
  return t
```
